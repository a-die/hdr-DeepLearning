```python
import pandas as pd
```

```python
with open('./smiles_20w.txt') as f:
    lines = f.readlines()
```

```python
smiles_20w = []
for smi in lines:
    smiles_20w.append(smi.split()[0])
```

```python
min(smiles_20w,key=len)
```

```python
for smi in smiles_20w:
    if len(smi) < 5:
        print(smi)
```
